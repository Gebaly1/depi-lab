```python
def find_Primenums(start,end):
  prims=[]

  for num in range(start,end+1):
    for i in range(2,num):
      if((num % i)==0):
        break
    else:
      prims.append(num)

  return prims



print(find_Primenums(2,40))
```

```python
class X :
  def __init__(self,name,age) -> None:
    self.name=name
    self.age=age


  def __str__(self) -> str:
    return f'{self.name} , {self.age} , '


class Z:
  def __init__(self,height,weight) -> None:
    self.height=height
    self.weight=weight


  def __str__(self) -> str:
    return f'{self.height} , {self.weight} '



class C(X , Z):
  def __init__(self, name, age,height,weight) -> None:
    X.__init__(self,name, age)
    Z.__init__(self,height, weight)


  def __str__(self) -> str:
    return f'{X.__str__(self)}{Z.__str__(self)}'



c1=C(name='hady',age='28',height='180',weight='100')

print(c1)




```

```python
class Video:                               # Define a class named 'Video'
    def __init__(self,video_codec):
        self.video_codec = 'H.264'

    def play_video(self):
        return 'Playing video...'

class Audio:                               # Define a class named 'Audio'
    def __init__(self,audio_codec):
        self.audio_codec = 'AAC'

    def play_audio(self):
        return 'Playing audio...'

class Multimedia(Video, Audio):            # Define a subclass named 'Multimedia' that inherits from 'Video' and 'Audio'
    def __init__(self):
        Video.__init__(self)               # Call the parent 'Video' class's '__init__' method
        Audio.__init__(self)               # Call the parent 'Audio' class's '__init__' method

    def play(self):
        return self.play_video() + ' ' + self.play_audio()   # Play both video and audio

multimedia = Multimedia()                  # Create an instance of the 'Multimedia' class
print(multimedia.play())                   # Output: 'Playing video... Playing audio...'
```

```python
def pangrams(s):
    st='abcdefghijklmnopqrstuvwxyz '
    for i in st:
      print(i)
      if i not in s.lower():
        print('----',i)
        return 'not pangram'
    else:
      return 'pangram'

print(pangrams('We promptly judged antique ivory buckles for the next prize'))
```

```python
def marsExploration(s):
    n=len(s)
    curs=3
    cnt=0
    sig='SOS'
    for i in range(0,n,3):
      cs=s[i:curs]
      if cs!=sig:
        for j in range(3):
          if(cs[j]!=sig[j]):
            cnt+=1
      curs+=3
    return cnt


print(marsExploration('SOSSOT'))
```
